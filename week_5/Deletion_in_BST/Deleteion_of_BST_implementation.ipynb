```python
class BinarySearchTreeNode:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.val = data


## function definition of insertion of a node in a given BST
def insertBST(root, value):
    if root is None:
        return BinarySearchTreeNode(value)
    else:
        # not allow any duplicates
        if root.val == value:
            return root
        elif root.val < value:
            root.right = insertBST(root.right, value)
        else:
            root.left = insertBST(root.left, value)
        return root
    

## function definition of searching of a element in a given BST
## best/avg case complexity : O(log n)
## worst case complexity (skewed binary search tree): O(n)
def searchBST(root, target):
    if root is None:
        return False
    elif root.val < target:
        return searchBST(root.right, target)
    elif root.val > target:
        return searchBST(root.left, target)
    else:
        return True


## function to get the minimum data in a BST
def minVal(node):
    current = node
    while current.left:
        current = current.left
    return current

## function to get the minimum data in a BST
def maxValue(node):
    current = node
    while current.right:
        current = current.right
    return current



## function to delete a node from a binary search tree
def deleteNode(root, key):
###   Right Sub Tree maximum data
    if root is None:
        return root
    
    ## 1. Search for a given key
    if root.val < key:
        ## recursion towards right
        root.right = deleteNode(root.right, key)
    elif root.val > key:
        root.left = deleteNode(root.left, key)
    else:
        ## easy case
        if root.left is None:
            return root.right
        if root.right is None:
            return root.left
        ## Difficult case
        tempNode = minVal(root.right)
        root.val = tempNode.val
        root.right = deleteNode(root.right, tempNode.val)
    return root
###   Left Sub Tree maximum data
    if root is None:
        return root
    
    ## 1. Search for a given key
    if root.val > key:
        ## recursion towards left
        root.left = deleteNode(root.left, key)
    elif root.val < key:
        root.right = deleteNode(root.right, key)
    else:
        ## easy case
        if root.right is None:
            return root.left
        if root.left is None:
            return root.right
        ## Difficult case
        tempNode = maxValue(root.left)
        root.val = tempNode.val
        root.left = deleteNode(root.left, tempNode.val)
    return root    



## function of inorder traversal(root)
def inOrderTraversal(root):
    if root:
        inOrderTraversal(root.left)
        print(root.val, end= " ")
        inOrderTraversal(root.right)


## Driver code
root = BinarySearchTreeNode(100)
root = insertBST(root, 80)
root = insertBST(root, 120)
root = insertBST(root, 70)
root = insertBST(root, 85)
root = insertBST(root, 110)
## function calling to display the result
inOrderTraversal(root)

#print(searchBST(root, 113))
key = 100
deleteNode(root, key)
print("After deletion of a node, the inorder traversal is: ")
inOrderTraversal(root)
```
